<h2>ChatBot for Food delivery services using Deep learning</h2>

```python
#Importing required libraries
import json 
import numpy as np 
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, GlobalAveragePooling1D
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder
```

```python
#Reading the intents file
with open('D:\G_DRIVE\Deeplearning\intents.json') as file:
    input_data=json.load(file)
```

```python
print(input_data)
```

```python
#Converting the json into required formats
training_sentences=[]
training_labels=[]
responses=[]
labels=[]

for i in input_data['intents']:
    for p in i['patterns']:
        training_sentences.append(p)
        training_labels.append(i['tag'])
    responses.append(i['responses'])
    
    if i['tag'] not in labels:
        labels.append(i['tag'])
    
```

```python
print(training_sentences)
print(training_labels)
print(responses)
print(labels)
```

```python
#Encoding the target labels
lbl = LabelEncoder()
lbl.fit(training_labels)
training_labels = lbl.transform(training_labels)
```

```python
#Defining certain measures
vocab_size = 1000 # Maximum no of unique words
embedding_dim = 16 # Each words will be represented as 16 dimension vectors
max_len = 20 # maximum length of longest document in the data
oov_token = "<OOV>" # Unkown words which are not in vocabulary will be replaced by this value

#Tokenizing
tokenizer = Tokenizer(num_words=vocab_size, oov_token=oov_token)
tokenizer.fit_on_texts(training_sentences)
```

```python
#Converting tokenized inputs to sequences
sequences = tokenizer.texts_to_sequences(training_sentences)
```

```python
#Padding the input sequences to make sure all are in same size
padded_sequences = pad_sequences(sequences, truncating='post', maxlen=max_len)
```

```python
#BDefining layers and compiling the neural network
model = Sequential()
model.add(Embedding(vocab_size, embedding_dim, input_length=max_len))
model.add(GlobalAveragePooling1D())
model.add(Dense(16, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(len(labels), activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy', 
              optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python
#Defining the epochs and training the model
epochs = 300
history = model.fit(padded_sequences, training_labels, epochs=epochs)
```

```python
#Saving the model, tokenizer and label encoders
model.save("chatbot_model")

import pickle

# to save the fitted tokenizer
with open('tokenizer.pickle', 'wb') as handle:
    pickle.dump(tokenizer, handle)
    
# to save the fitted label encoder
with open('label_encoder.pickle', 'wb') as ecn_file:
    pickle.dump(lbl, ecn_file)
```
